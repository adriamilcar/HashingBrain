## Discretization of input space

```python
from torchvision import transforms, models
from torch.utils.data import DataLoader, Dataset

class TensorDataset(Dataset):
    def __init__(self, images):
        self.images = images
        self.transform = transforms.Compose([
            transforms.ToPILImage(),
            #transforms.Resize((224, 224)),
            transforms.Resize((299, 299)),  # Adjusted size for Inception
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        image = self.transform(image)
        return image

dataloader = DataLoader(TensorDataset(torch.tensor(dataset).permute(0, 3, 1, 2)), batch_size=64, shuffle=False)

# Load pre-trained VGG-16
#pretrained_model = models.vgg16(pretrained=True)
#pretrained_model.eval()
#pretrained_model.classifier[6] = torch.nn.Identity()

# Load pre-trained Inception V3
pretrained_model = models.inception_v3(pretrained=True, aux_logits=True)
pretrained_model.eval()
pretrained_model.fc = torch.nn.Identity()
```

```python
# Extract features
features = []
with torch.no_grad():
    for images in dataloader:
        outputs = pretrained_model(images)
        features.extend(outputs.cpu().numpy())

features = np.array(features)
```

```python
from sklearn.manifold import TSNE

# Apply t-SNE
tsne = TSNE(n_components=2)
images_2d = tsne.fit_transform(features)
```

```python
from sklearn.manifold import MDS

# Apply MDS
mds = MDS(n_components=2)
images_2d = mds.fit_transform(features)
```

```python
images_2d = np.load('results/2d_imageSpace_InceptionV3-MDS.npy')
```

```python
from scipy.spatial import ConvexHull
from shapely.geometry import Polygon
from shapely.ops import unary_union
from itertools import combinations

plt.figure(figsize=(6, 6))

n_active_units = np.count_nonzero(np.any(embeddings, axis=0))
#n_active_units = 20

cmap = plt.get_cmap('gist_rainbow')
colors = [cmap(val) for val in np.linspace(0, 1, n_active_units)]

hull_polygons = []
for i in range(n_active_units):
    img_indxs = indxs_embeddings_active_specific[i][0]
    if len(img_indxs) > 3:
        plt.scatter(images_2d[img_indxs, 0], images_2d[img_indxs, 1], s=50, c=colors[i])
        
        hull = ConvexHull(images_2d[img_indxs])
        hull_polygon = Polygon(images_2d[img_indxs][hull.vertices])
        hull_polygons.append(hull_polygon)
        
        for simplex in hull.simplices:
            plt.plot(images_2d[img_indxs][simplex, 0], images_2d[img_indxs][simplex, 1], c=colors[i])
        
        plt.fill(images_2d[img_indxs][hull.vertices, 0], images_2d[img_indxs][hull.vertices, 1], c=colors[i], alpha=0.3)

# Calculate the average of the intersection over union for all pairs of hulls
total_intersection_over_union = 0
pair_count = 0

for i, j in combinations(range(len(hull_polygons)), 2):
    intersection = hull_polygons[i].intersection(hull_polygons[j]).area
    union = hull_polygons[i].union(hull_polygons[j]).area
    total_intersection_over_union += intersection / union if union > 0 else 0
    pair_count += 1

# Compute the average overlap metric
overlap_metric = total_intersection_over_union / pair_count if pair_count > 0 else 0

plt.title(f'Overlap: {overlap_metric:.2f}', fontsize=18)
plt.axis('off')
plt.tight_layout()
plt.show()
```

### Hulls based on clustering in 2D space

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
from sklearn.cluster import DBSCAN
from shapely.geometry import Polygon
from itertools import combinations

plt.figure(figsize=(6, 6))

n_active_units = np.count_nonzero(np.any(embeddings, axis=0))
#n_active_units = 10

cmap = plt.get_cmap('gist_rainbow')
colors = [cmap(val) for val in np.linspace(0, 1, n_active_units)]

hull_polygons = []
for i in range(n_active_units):
    img_indxs = indxs_embeddings_active_specific[i][0]
    points = images_2d[img_indxs]
    
    # Apply DBSCAN
    clustering = DBSCAN(eps=1, min_samples=5).fit(points)
    labels = clustering.labels_
    #print(labels)

    # Process each cluster
    for k in set(labels):
        if k != -1:  # Ignore noise points
            class_member_mask = (labels == k)
            cluster_points = points[class_member_mask]
            
            if len(cluster_points) >= 3:
                try:
                    hull = ConvexHull(cluster_points)
                    polygon = Polygon(cluster_points[hull.vertices])
                    hull_polygons.append(polygon)

                    # Plot the convex hull
                    x, y = polygon.exterior.xy
                    plt.fill(x, y, c=colors[i], alpha=0.3)
                except Exception as e:
                    print(f"Could not create hull for cluster {k}: {e}")

# Calculate the average of the intersection over union for all pairs of hulls
total_intersection_over_union = 0
pair_count = 0

for i, j in combinations(range(len(hull_polygons)), 2):
    intersection = hull_polygons[i].intersection(hull_polygons[j]).area
    union = hull_polygons[i].union(hull_polygons[j]).area
    total_intersection_over_union += intersection / union if union > 0 else 0
    pair_count += 1

# Compute the average overlap metric
average_overlap_metric = total_intersection_over_union / pair_count if pair_count > 0 else 0

plt.title(f'Overlap: {average_overlap_metric:.3f}', fontsize=18)
plt.axis('off')
plt.tight_layout()
plt.savefig('plots/overlap_imageSpace_ortho.png', dpi=600)
plt.show()
```
