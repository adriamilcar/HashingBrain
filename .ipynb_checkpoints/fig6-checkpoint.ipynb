```python
%matplotlib inline
from utils import *
from plotting import *
from models import *
import os
import pickle
import seaborn as sb
```

```python
tasks = ['doubleTmaze', 'cylinder', 'permanence', 'thorndike']
model_types = ['sparse-AE']
```

## Results

```python
results = {}
for model_type in model_types:
    results[model_type] = {}
    for task_train in tasks:
        print(f"Evaluating models trained on {task_train} for model type {model_type}")
        
        # Load the model trained on the current task
        model = Conv_AE(n_hidden=1000).to('cuda')
        model.load_state_dict(torch.load(f'saved_models/main_{task_train}_{model_type}_1000hidden.pth'))
        
        for task_test in tasks:
            print(f"Testing on {task_test}")
            
            # Load the test dataset
            dataset, position = load_dataset(directory=f'datasets/{task_test}', task=task_test)
            
            # Get latent embeddings
            embeddings = get_latent_vectors(dataset, model)
            embeddings[embeddings < 1e-5] = 0
            
            # Generate ratemaps and cleaned ratemaps
            ratemaps = generate_ratemaps(embeddings, position, n_bins=60, filter_width=4, n_bins_padding=10)
            cleaned_ratemaps = clean_ratemaps(ratemaps)
            
            # Place field density
            all_num_fields, all_centroids, all_sizes = stats_place_fields(cleaned_ratemaps, min_pix_cluster=0.03)
            place_field_density = np.histogram(all_num_fields, bins=np.max(all_num_fields) + 1, density=True)[0]

            # Average spatial information
            unfiltered_ratemaps = generate_ratemaps(embeddings, position, n_bins=30, filter_width=0, n_bins_padding=0)
            cleaned_unfiltered_ratemaps = clean_ratemaps(unfiltered_ratemaps)
            occupancy_map_norm = generate_occupancy_map(position, n_bins=30, filter_width=0, n_bins_padding=0, norm=True)
            avg_spatial_info = np.mean(spatial_information(cleaned_unfiltered_ratemaps, occupancy_map_norm))
        
            # Linear decoding error
            ratemaps_reshaped = ratemaps.reshape((ratemaps.shape[0], int(ratemaps.shape[1]*ratemaps.shape[2]))).T
            pos_values = np.linspace(np.min(position), np.max(position), ratemaps.shape[1])
            position_for_ratemaps = np.array([[i, j] for i in pos_values for j in pos_values])
            norm_factor = (np.max(position) - np.min(position)) * np.sqrt(2)
            mean_decoding_error_ratemaps = linear_decoding_error(ratemaps_reshaped, position_for_ratemaps, norm=norm_factor)
            
            # Store the results in the dictionary
            results[model_type].setdefault(task_train, {})[task_test] = {
                'ratemaps': cleaned_ratemaps,
                'place_field_density_hist': place_field_density,
                'avg_spatial_info': avg_spatial_info,
                'mean_decoding_error_ratemaps': mean_decoding_error_ratemaps
            }
```

```python
# Optionally save the results to a file
with open('results/cross_task_evaluation.pkl', 'wb') as f:
    pickle.dump(results, f)
```

## Plots

```python
# Load the sparse (train) data
file_path = os.path.join('results', 'spatial_allModels.pkl')
with open(file_path, 'rb') as f:
    results_train = pickle.load(f)

# Custom style settings
colors = {'sparse-AE': '#4E79A7'}
edge_color = '#ffffff'
alpha_value = 0.8

# Extract and pool histograms for the 'sparse-AE' model type across all tasks and models for test
pooled_histograms_test = []
for task_train in tasks:
    for task_test in tasks:
        if task_test != task_train:
            pooled_histograms_test.append(results['sparse-AE'][task_train][task_test]['place_field_density_hist'])

# Extract and pool histograms for the 'sparse-AE' model type across all tasks and models for train
pooled_histograms_train = []
for task in tasks:
    pooled_histograms_train.append(results_train['sparse-AE'][task]['place_field_density_hist'])

# Find the maximum length of the histograms
max_length = max(max(len(hist) for hist in pooled_histograms_test), max(len(hist) for hist in pooled_histograms_train))

# Pad histograms with zeros to make them the same length
padded_histograms_test = np.array([np.pad(hist, (0, max_length - len(hist)), 'constant') for hist in pooled_histograms_test])
padded_histograms_train = np.array([np.pad(hist, (0, max_length - len(hist)), 'constant') for hist in pooled_histograms_train])

# Collapse bins above 4 place fields into a single bin
collapsed_histograms_test = np.zeros((len(padded_histograms_test), 5))
collapsed_histograms_train = np.zeros((len(padded_histograms_train), 5))
collapsed_histograms_test[:, :4] = padded_histograms_test[:, :4]
collapsed_histograms_test[:, 4] = padded_histograms_test[:, 4:].sum(axis=1)
collapsed_histograms_train[:, :4] = padded_histograms_train[:, :4]
collapsed_histograms_train[:, 4] = padded_histograms_train[:, 4:].sum(axis=1)

# Compute the mean and standard deviation across the histograms
mean_counts_test = np.mean(collapsed_histograms_test, axis=0)
std_counts_test = np.std(collapsed_histograms_test, axis=0)
mean_counts_train = np.mean(collapsed_histograms_train, axis=0)
std_counts_train = np.std(collapsed_histograms_train, axis=0)

# Normalize the mean counts to sum to 1 (as probabilities)
mean_counts_test /= mean_counts_test.sum()
mean_counts_train /= mean_counts_train.sum()

# Create the bar plots with error bars
fig, axs = plt.subplots(1, 2, figsize=(10, 6), sharey=True)

# Plot for sparse (train)
bars_train = axs[0].bar(np.arange(5), mean_counts_train, yerr=std_counts_train, width=1, color=colors['sparse-AE'], alpha=alpha_value / 2, edgecolor=edge_color, capsize=5)
axs[0].set_xlabel('# place fields', fontsize=20, fontweight='bold')
axs[0].set_ylabel('prob.', fontsize=20, fontweight='bold')
axs[0].set_yticks(np.linspace(0, 1, 6))
axs[0].set_yticklabels(np.linspace(0, 1, 6).round(1), fontsize=18)
axs[0].set_xticks(np.arange(5))
axs[0].set_xticklabels([0, 1, 2, 3, '4+'], fontsize=18)
axs[0].set_ylim(0, 1)
axs[0].set_title('same env.', fontsize=20, fontweight='bold')

# Plot for sparse (test)
bars_test = axs[1].bar(np.arange(5), mean_counts_test, yerr=std_counts_test, width=1, color=colors['sparse-AE'], alpha=alpha_value, edgecolor=edge_color, capsize=5)
axs[1].set_xlabel('# place fields', fontsize=20, fontweight='bold')
axs[1].set_yticks(np.linspace(0, 1, 6))
axs[1].set_yticklabels(np.linspace(0, 1, 6).round(1), fontsize=18)
axs[1].set_xticks(np.arange(5))
axs[1].set_xticklabels([0, 1, 2, 3, '4+'], fontsize=18)
axs[1].set_ylim(0, 1)
axs[1].set_title('across envs.', fontsize=20, fontweight='bold')

# Add gridlines
for ax in axs:
    ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Despine and customize tick params
for ax in axs:
    sb.despine(ax=ax, left=True, bottom=True)
    ax.tick_params(axis='y', labelsize=18)

# Overall layout adjustments
plt.tight_layout()

plt.savefig('plots/fig6/A_placeFieldDistr_sparse_train_test.svg', dpi=600)

plt.show()
```

```python
# Load the results (assuming they are saved in the previous step)
with open('results/cross_task_evaluation.pkl', 'rb') as f:
    results = pickle.load(f)

# Extract and pool data for the 'sparse-AE' model type across all tasks and models for train and test
spatial_info_sparse_test = []
mean_decoding_error_ratemaps_sparse_test = []
spatial_info_sparse_train = []
mean_decoding_error_ratemaps_sparse_train = []

for task_train in tasks:
    for task_test in tasks:
        if task_test != task_train:
            spatial_info_sparse_test.append(results['sparse-AE'][task_train][task_test]['avg_spatial_info'])
            mean_decoding_error_ratemaps_sparse_test.append(results['sparse-AE'][task_train][task_test]['mean_decoding_error_ratemaps'])
        else:
            spatial_info_sparse_train.append(results['sparse-AE'][task_train][task_test]['avg_spatial_info'])
            mean_decoding_error_ratemaps_sparse_train.append(results['sparse-AE'][task_train][task_test]['mean_decoding_error_ratemaps'])

# Custom style settings
colors = {'sparse': '#4E79A7'}
edge_color = '#ffffff'
alpha_value = 0.8
bar_width = 0.5  # Narrower bars

# Compute means and standard deviations for pooled data
mean_train_spatial = np.mean(spatial_info_sparse_train)
std_train_spatial = np.std(spatial_info_sparse_train)

mean_test_spatial = np.mean(spatial_info_sparse_test)
std_test_spatial = np.std(spatial_info_sparse_test)

mean_train_rate = np.mean(mean_decoding_error_ratemaps_sparse_train)
std_train_rate = np.std(mean_decoding_error_ratemaps_sparse_train)

mean_test_rate = np.mean(mean_decoding_error_ratemaps_sparse_test)
std_test_rate = np.std(mean_decoding_error_ratemaps_sparse_test)

# Plot
fig, axs = plt.subplots(1, 2, figsize=(10, 6), sharey=False)

# Spatial info plot
axs[0].bar('same env.', mean_train_spatial, yerr=std_train_spatial, color=colors['sparse'], alpha=alpha_value / 2, edgecolor=edge_color, capsize=5, width=bar_width)
axs[0].bar('across envs.', mean_test_spatial, yerr=std_test_spatial, color=colors['sparse'], alpha=alpha_value, edgecolor=edge_color, capsize=5, width=bar_width)
axs[0].set_ylabel('spatial info. per neuron\n(bit/sample)', fontsize=20, fontweight='bold')
axs[0].set_xticks([0, 1])
axs[0].set_ylim(-0.01)
axs[0].set_xticklabels(['same env.', 'across envs.'], fontsize=18, fontweight='bold')
axs[0].tick_params(axis='y', labelsize=18)
axs[0].grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Mean decoding error ratemaps plot
axs[1].bar('same env.', mean_train_rate, yerr=std_train_rate, color=colors['sparse'], alpha=alpha_value / 2, edgecolor=edge_color, capsize=5, width=bar_width)
axs[1].bar('across envs.', mean_test_rate, yerr=std_test_rate, color=colors['sparse'], alpha=alpha_value, edgecolor=edge_color, capsize=5, width=bar_width)
axs[1].set_ylabel('mean position error\n(% of max. distance)', fontsize=20, fontweight='bold')
axs[1].set_xticks([0, 1])
axs[1].set_ylim(-0.01,1.01)
axs[1].set_xticklabels(['same env.', 'across envs.'], fontsize=18, fontweight='bold')
axs[1].tick_params(axis='y', labelsize=18)
axs[1].grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Despine for a cleaner look
for ax in axs:
    sb.despine(ax=ax, left=True, bottom=True)

# Adjust the space between subplots
plt.subplots_adjust(wspace=0.5)

plt.tight_layout()

plt.savefig('plots/fig6/B_info_and_decoding_ratemaps_sparse_train_test.svg', dpi=600)

plt.show()
```
